1. What is a random variable in probability theory?

A. A random variable in probability theory is a function that assigns numerical values to the outcomes of a random experiment. It helps quantify uncertainty and can be either discrete (taking countable values like whole numbers) or continuous (taking values from a range or interval). For example, the number rolled on a die or the time taken for an event are random variables.

2. What are the types of random variables?

A. Random variables are of two types: discrete and continuous. A discrete random variable takes countable values, like the number of heads in coin tosses, and is described using a probability mass function (PMF). A continuous random variable takes values from an uncountable range, like time or height, and is described using a probability density function (PDF).

3. What is the difference between discrete and continuous distributions?

A.A discrete distribution deals with countable outcomes, like rolling a die, and uses a Probability Mass Function (PMF). A continuous distribution handles uncountable outcomes over an interval, like measuring time or height, and uses a Probability Density Function (PDF). Discrete gives probabilities at specific points, while continuous gives probabilities over intervals.

4. What are probability distribution functions (PDF)?

A. A Probability Density Function (PDF) is used for continuous random variables to show how values are distributed. It gives the relative likelihood of the variable falling within a range, not at a specific point. The area under the curve of the PDF over an interval represents the probability, and the total area under the curve is always 1.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

A. A Cumulative Distribution Function (CDF) gives the probability that a random variable is less than or equal to a certain value, while a Probability Density Function (PDF) shows the likelihood of the variable falling within a small range. For continuous variables, the PDF is the derivative of the CDF, and the CDF is the integral of the PDF. In short, the PDF shows density, and the CDF shows accumulated probability up to a point.


6. What is a discrete uniform distribution?

A. A discrete uniform distribution is a distribution where all possible distinct outcomes have the same probability. It applies to situations with a finite number of equally likely outcomes, such as tossing a fair coin or rolling a fair die. Each outcome has a probability of 1/N , where n is the total number of outcomes.

7. What are the key properties of a Bernoulli distribution?

A. The Bernoulli distribution models a single trial with only two outcomes: success (1) with probability p, and failure (0) with probability 1-p. It has a mean of p and a variance ofp(1-p), and is used to represent binary outcomes like yes/no or true/false.

8. What is the binomial distribution, and how is it used in probability?

A. The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success p. It is defined by two parameters:
n (number of trials) and p (probability of success). It is used in probability to calculate the likelihood of getting a certain number of successes out of n trials, such as getting 3 heads in 5 coin tosses.

9. What is the Poisson distribution and where is it applied?

A. The Poisson distribution models the probability of a given number of events occurring in a fixed interval of time or space, when the events happen independently and at a constant average rate Œª. It is used to model rare or random events, such as the number of phone calls at a call center per hour, or the number of accidents at a traffic intersection per day.

17. What are point estimates and interval estimates in statistics?

A. In statistics, a point estimate gives a single best guess of a population parameter (like the mean or proportion) based on sample data‚Äîfor example, using the sample mean to estimate the population mean.

An interval estimate, on the other hand, provides a range of values (called a confidence interval) within which the parameter is likely to lie, along with a level of confidence (e.g., 95%). It accounts for sampling variability and gives more information than a point estimate alone.

18. What is the significance of confidence intervals in statistical analysis?

A. Confidence intervals are important in statistical analysis because they provide a range of values within which a population parameter is likely to lie, along with a confidence level (e.g., 95%). Unlike point estimates, which give a single value, confidence intervals reflect the uncertainty and variability inherent in sample data. They help assess the precision of an estimate and support better, more informed decision-making.

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the discrete uniform distribution
low = 1   # minimum value (inclusive)
high = 7  # maximum value (exclusive for scipy randint, so 7 for a die)

# Create a discrete uniform distribution object
dist = randint(low, high)

# Define x values (possible outcomes)
x = np.arange(low, high)

# Compute PMF values
pmf_values = dist.pmf(x)

# Plot the PMF
plt.stem(x, pmf_values)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution (PMF) - Fair Die')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: Probability of the outcome x
    """
    if x not in [0, 1]:
        return 0  # Bernoulli is only defined for x = 0 or 1
    return p if x == 1 else (1 - p)

# Example usage
p = 0.7  # Probability of success
print("P(X=0):", bernoulli_pdf(0, p))
print("P(X=1):", bernoulli_pdf(1, p))
```

```python
#4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10       # Number of trials
p = 0.5      # Probability of success
size = 1000  # Number of simulated samples

# Simulate binomial distribution
samples = np.random.binomial(n, p, size)

# Plot histogram of simulated values
plt.hist(samples, bins=range(n+2), align='left', edgecolor='black', density=True, alpha=0.7, label='Simulated')

# Overlay the theoretical PMF
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'o-', color='red', label='Theoretical PMF')

# Labels and title
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(x)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```

```python
#5. Create a Poisson distribution and visualize it using Python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set Poisson distribution parameter (mean rate of occurrence)
lambda_ = 4  # Average number of events (e.g., 4 per interval)

# Generate x values (possible number of events)
x = np.arange(0, 15)

# Calculate PMF for each x
pmf = poisson.pmf(x, mu=lambda_)

# Plot the PMF
plt.stem(x, pmf, basefmt=" ", linefmt='blue', markerfmt='bo')
plt.title(f'Poisson Distribution (Œª = {lambda_})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for discrete uniform distribution
low = 1       # Minimum value (inclusive)
high = 7      # Maximum value (exclusive in scipy, so use 7 for 1 to 6)
dist = randint(low, high)

# Possible outcome values
x = np.arange(low - 1, high + 1)

# Calculate CDF values
cdf_values = dist.cdf(x)

# Plot the CDF
plt.step(x, cdf_values, where='post', label='CDF', color='green')
plt.title('CDF of Discrete Uniform Distribution (Die Roll: 1 to 6)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(np.arange(low - 1, high + 1))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#7. Generate a continuous uniform distribution using NumPy and visualize it
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters for the uniform distribution
a = 0     # Lower bound
b = 1     # Upper bound
size = 1000  # Number of samples

# Generate random samples from a continuous uniform distribution
samples = np.random.uniform(a, b, size)

# Create x values and compute the PDF
x = np.linspace(a, b, 100)
pdf = uniform.pdf(x, loc=a, scale=b-a)

# Plot histogram of samples and overlay the PDF
plt.hist(samples, bins=20, density=True, alpha=0.6, edgecolor='black', label='Sample Histogram')
plt.plot(x, pdf, 'r-', lw=2, label='Theoretical PDF')

# Labels and title
plt.title('Continuous Uniform Distribution (a=0, b=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#8. Simulate data from a normal distribution and plot its histogram
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0        # Mean
sigma = 1     # Standard deviation
size = 1000   # Number of samples

# Simulate data
samples = np.random.normal(mu, sigma, size)

# Create x values and compute the theoretical PDF
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Plot histogram of samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Sample Histogram')

# Overlay the theoretical normal distribution
plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

# Labels and title
plt.title('Normal Distribution (Œº=0, œÉ=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from the dataset and plot them.

    Parameters:
    data (list or np.array): Input numerical data
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot 1: Line plot of Z-scores
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(y=0, color='red', linestyle='--')
    plt.title('Z-Scores Line Plot')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.grid(True)

    # Plot 2: Histogram of Z-scores
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=10, edgecolor='black', color='skyblue', density=True)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage
data = [45, 50, 55, 60, 65, 70, 75, 80, 85]
z_scores = calculate_and_plot_z_scores(data)
```

```python
#10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution
import numpy as np
import matplotlib.pyplot as plt

def demonstrate_clt(sample_size=30, num_samples=1000):
    """
    Demonstrate the Central Limit Theorem using samples from an exponential distribution.

    Parameters:
    sample_size (int): Number of observations in each sample
    num_samples (int): Number of samples to draw
    """
    # Population: exponential distribution (non-normal)
    population = np.random.exponential(scale=2.0, size=100000)

    # Draw multiple samples and compute sample means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))

    # Plot histogram of sample means
    plt.hist(sample_means, bins=30, density=True, edgecolor='black', color='skyblue', alpha=0.7)
    plt.title(f'CLT Demonstration (Sample Size = {sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
demonstrate_clt(sample_size=30, num_samples=1000)
```
