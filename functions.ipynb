A. A function is typically defined using the def keyword followed by a function name, parameters, and a block of code. 4. What is the purpose of the `return` statement in a Python function?

A. The return statement in a Python function is used to send a result back to the caller of the function. It allows the function to produce an output that can be stored, printed, or used in further calculations.

5. What are iterators in Python and how do they differ from iterables?

A. An iterable is any object that can be looped over (e.g., with a for loop). Common examples include lists, tuples, strings, and sets. Internally, an iterable must implement the __iter__() method, which returns an iterator.
An iterator is an object that represents a stream of data and returns items one at a time using the __next__() method. It also implements the __iter__() method, meaning every iterator is also an iterable, but not every iterable is an iterator.

8. What is a lambda function in Python and when is it typically used?

A. A lambda function in Python is a small, anonymous function defined using the lambda keyword instead of def. It can take any number of arguments but must contain only a single expression, which is evaluated and returned. Lambda functions are typically used for short, simple operations where defining a full function would be unnecessarily verbose.

9. Explain the purpose and usage of the `map()` function in Python.

A. The map() function in Python is used to apply a given function to each item in an iterable (such as a list or tuple) and return a new iterator with the results. Its purpose is to simplify the process of transforming data without writing explicit loops.

10. 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

A.map() applies a given function to each item of an iterable and returns an iterator with the results. It is used for transforming data.
Example: map(lambda x: x * 2, [1, 2, 3]) → [2, 4, 6]
filter() applies a function that returns True or False to each item and keeps only the items for which the function returns True. It is used for filtering data.
Example: filter(lambda x: x % 2 == 0, [1, 2, 3, 4]) → [2, 4]
reduce() (from the functools module) applies a function cumulatively to the items of an iterable, reducing the iterable to a single value. It is used for aggregation.
Example: reduce(lambda x, y: x + y, [1, 2, 3, 4]) → 10

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
# the list.
def sum_of_even_numbers(numbers):
    even_sum = sum(num for num in numbers if num % 2 == 0)
    return even_sum

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number.")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
for power in powers_of_two(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Yield each line, removing any extra newline characters
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"Error: {e}")

# Example usage:
file_path = input("Enter the path to the file: ")
for line in read_file_line_by_line(file_path):
    print(line)


```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
tuples_list = [(1, 5), (3, 1), (4, 8), (2, 0)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print("Sorted list based on second element:", sorted_list)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32
celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, s))
input_string = "Hello, World!"
result = remove_vowels(input_string)
print("String without vowels:", result)
```
