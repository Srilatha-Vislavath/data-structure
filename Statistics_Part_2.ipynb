1. What is hypothesis testing in statistics

A. Hypothesis testing is a statistical technique used to make decisions or draw conclusions about a population based on sample data. It involves forming a null hypothesis (Hâ‚€), which represents no effect or difference, and an alternative hypothesis (Hâ‚), which represents the presence of an effect or difference. By analyzing the data using test statistics and p-values, we determine whether to reject the null hypothesis in favor of the alternative.


2. What is the null hypothesis, and how does it differ from the alternative hypothesis

A. In hypothesis testing, the **null hypothesis (Hâ‚€)** assumes no effect or difference, serving as the default assumption, while the **alternative hypothesis (Hâ‚)** suggests there is a significant effect or difference. The test evaluates sample data to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative.


3.  What is the significance level in hypothesis testing, and why is it important

A. The **significance level** (denoted as **Î±**) in hypothesis testing is the threshold for deciding whether to reject the null hypothesis. Common values are **0.05**, **0.01**, or **0.10**, which represent a **5%**, **1%**, or **10%** risk of rejecting the null hypothesis when it is actually true (Type I error). It is important because it sets the standard for how much evidence is required before concluding that an effect is statistically significant. A lower Î± reduces the chance of a false positive but increases the chance of missing a real effect (Type II error).


4. What does a P-value represent in hypothesis testing

A. In hypothesis testing, a **p-value** represents the probability of obtaining a test result **at least as extreme** as the one observed, **assuming the null hypothesis is true**.

A **small p-value** (typically less than the significance level, e.g., 0.05) indicates that the observed data is unlikely under the null hypothesis, suggesting evidence **against Hâ‚€**.
A **large p-value** means the data is consistent with Hâ‚€, so we **fail to reject** it.

In short, the p-value helps decide whether to reject the null hypothesis based on how surprising the data is under the assumption that Hâ‚€ is true.


5.  How do you interpret the P-value in hypothesis testing

A. In hypothesis testing, the **p-value** helps determine the strength of evidence against the null hypothesis (Hâ‚€):

* If **p-value â‰¤ Î±** (e.g., 0.05), **reject Hâ‚€**: There is **strong evidence** that the effect is real or significant.
* If **p-value > Î±**, **fail to reject Hâ‚€**: There is **not enough evidence** to support the alternative hypothesis.

In short, a **small p-value** suggests the observed result is unlikely under Hâ‚€, while a **large p-value** means the result is likely due to chance.


6. What are Type 1 and Type 2 errors in hypothesis testing

A. In hypothesis testing, a **Type 1 error** occurs when we reject a true null hypothesis, meaning we detect an effect that doesn't actually exist (false positive). A **Type 2 error** happens when we fail to reject a false null hypothesis, meaning we miss a real effect (false negative). Type 1 errors are controlled by the significance level (Î±), while Type 2 errors relate to the testâ€™s power.


7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing

A. A **one-tailed test** checks for an effect in **only one direction** (either greater than or less than a certain value), while a **two-tailed test** checks for an effect in **both directions** (either significantly higher or lower than a certain value).

For example, if you're testing whether a new drug performs **better** than the old one, you'd use a **one-tailed test**. But if you're testing whether it performs **differently** (either better or worse), you'd use a **two-tailed test**.


8. What is the Z-test, and when is it used in hypothesis testing

A. A **Z-test** is a statistical method used in hypothesis testing to determine whether a sample mean significantly differs from a known population mean when the population standard deviation is known and the sample size is large (usually $n \geq 30$). It uses the Z-score to assess how far the sample statistic is from the population parameter under the null hypothesis.


9. How do you calculate the Z-score, and what does it represent in hypothesis testing

A. The **Z-score** is calculated using the formula:

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $\mu$ = population mean
* $\sigma$ = population standard deviation
* $n$ = sample size

In hypothesis testing, the Z-score tells us how many standard errors the sample mean is away from the population mean. A high absolute Z-score indicates the sample result is unlikely under the null hypothesis, guiding the decision to reject or not reject it.


10. What is the T-distribution, and when should it be used instead of the normal distribution

A. The **T-distribution** is a probability distribution used in statistics when estimating population parameters with **small sample sizes** (typically $n < 30$) and when the **population standard deviation is unknown**. It is similar to the normal distribution but has **heavier tails**, which provide a more conservative estimate of variability.

You should use the T-distribution instead of the normal distribution when dealing with small samples or when the population standard deviation is not known. As the sample size increases, the T-distribution approaches the normal distribution.


11. What is the difference between a Z-test and a T-test

A. The **Z-test** and **T-test** are both used in hypothesis testing to compare means, but they differ mainly in sample size and variance knowledge:

* **Z-test** is used when the **sample size is large** (typically $n \geq 30$) and the **population standard deviation is known**.
* **T-test** is used when the **sample size is small** ($n < 30$) and the **population standard deviation is unknown**.

In short, use a **Z-test** for large, well-known populations and a **T-test** for small samples or when population data is uncertain.


12. What is the T-test, and how is it used in hypothesis testing

A. A **T-test** is a statistical test used to determine whether there is a significant difference between the means of one or two groups, especially when the sample size is small and the population standard deviation is unknown. It helps assess whether observed differences are likely due to chance or represent a real effect.


13. What is the relationship between Z-test and T-test in hypothesis testing

A. The **Z-test** and **T-test** are both used to compare means in hypothesis testing, but they differ based on sample size and knowledge of population standard deviation. A **Z-test** is used when the sample size is large (typically â‰¥ 30) and the population standard deviation is known. A **T-test** is used for small samples and when the population standard deviation is unknown. Both tests assess how far the sample mean is from the population mean, but the **T-test accounts for more variability**, making it better suited for smaller or less certain datasets. As sample size increases, the T-test results approach those of the Z-test.


14. What is a confidence interval, and how is it used to interpret statistical results

A. A **confidence interval (CI)** is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean) with a certain level of confidence (commonly 95% or 99%).

It is used to interpret statistical results by giving an estimated range that is likely to include the true value. For example, a 95% confidence interval means that if the same study were repeated many times, 95% of the calculated intervals would contain the true population parameter. It provides more information than a single point estimate by expressing the **uncertainty and precision** of the estimate.


15. What is the margin of error, and how does it affect the confidence interval

A. The **margin of error** is the maximum expected difference between a sample statistic (like the sample mean) and the true population parameter. It reflects the **precision** of an estimate and is influenced by the **sample size**, **variability**, and **confidence level**.

In a confidence interval, the margin of error determines the **width** of the interval:

$$
\text{Confidence Interval} = \text{Point Estimate} \pm \text{Margin of Error}
$$

A **larger margin of error** results in a **wider interval** (more uncertainty), while a **smaller margin of error** gives a **narrower interval** (greater precision). Increasing the confidence level or decreasing the sample size usually increases the margin of error.


16. How is Bayes' Theorem used in statistics, and what is its significance

A. **Bayes' Theorem** is a statistical formula that helps update the probability of a hypothesis based on new evidence. It combines prior knowledge with observed data to calculate a revised (posterior) probability. It is widely used in fields like medical diagnosis, machine learning, and decision-making under uncertainty.


17. What is the Chi-square distribution, and when is it used

A. The **Chi-square distribution** is a statistical distribution used primarily for testing relationships between categorical variables. It's commonly applied in **goodness-of-fit** tests and **tests of independence** to compare observed and expected frequencies. The distribution is skewed to the right and only takes non-negative values, becoming more symmetrical with larger degrees of freedom.


18. What is the Chi-square goodness of fit test, and how is it applied

A. The **Chi-square goodness of fit test** is used to determine whether the observed frequencies of a single categorical variable match the expected frequencies based on a theoretical distribution. It helps assess how well the data fits a specific model, such as checking if a die is fair. If the difference between observed and expected values is large, the null hypothesis (that the data fits the expected distribution) is rejected.


19. What is the F-distribution, and when is it used in hypothesis testing

A. The **F-distribution** is a right-skewed probability distribution used primarily in hypothesis testing to compare **variances** between two populations. It is most commonly applied in **ANOVA (Analysis of Variance)** and **F-tests** to determine whether group means differ significantly or whether two datasets have equal variances. The shape of the F-distribution depends on two degrees of freedomâ€”one for the numerator and one for the denominatorâ€”and it is only defined for non-negative values.


20. What is an ANOVA test, and what are its assumptions

A. ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if at least one group differs significantly. It assumes the data is normally distributed, samples are independent, and group variances are equal. ANOVA helps identify differences in means without conducting multiple t-tests, reducing the risk of error.


21. What are the different types of ANOVA tests

A. The three main types of ANOVA are **One-Way ANOVA**, which compares means across three or more independent groups; **Two-Way ANOVA**, which assesses the effects of two independent variables and their interaction; and **Repeated Measures ANOVA**, used when the same subjects are measured multiple times under different conditions. Each type helps determine if group means differ significantly based on the study design.


22. What is the F-test, and how does it relate to hypothesis testing

A. The **F-test** is a statistical test used to compare two variances to determine if they are significantly different. It plays a key role in **hypothesis testing**, especially in **ANOVA (Analysis of Variance)**, where it helps test whether group means differ by analyzing the ratio of variances between groups and within groups. A large F-value suggests that the group means are not all equal, providing evidence to reject the null hypothesis. The F-test assumes that the data is normally distributed and the samples are independent.

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Step 1: Define parameters
low = 1     # Lower bound (inclusive)
high = 7    # Upper bound (exclusive)
size = 1000  # Number of samples

# Step 2: Generate random samples from a discrete uniform distribution
samples = randint.rvs(low, high, size=size)

# Step 3: Define values and compute PMF
x = np.arange(low, high)
pmf_values = randint.pmf(x, low, high)

# Step 4: Plot PMF
plt.figure(figsize=(8, 4))
plt.stem(x, pmf_values)
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True, linestyle="--", alpha=0.5)
plt.xticks(x)
plt.ylim(0, max(pmf_values) + 0.05)
plt.tight_layout()
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    - x: 0 or 1 (the outcome)
    - p: probability of success (0 â‰¤ p â‰¤ 1)

    Returns:
    - Probability value for the given x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage:
p = 0.7  # probability of success
print(f"P(X=1): {bernoulli_pdf(1, p)}")  # Success
print(f"P(X=0): {bernoulli_pdf(0, p)}")  # Failure
```

```python
#6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Step 1: Define the discrete uniform distribution range
low = 1     # inclusive
high = 7    # exclusive (so values 1 to 6)

# Step 2: Generate the x-values (possible outcomes)
x = np.arange(low, high + 1)

# Step 3: Compute the CDF for each value
cdf_values = randint.cdf(x, low, high)

# Step 4: Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='post', label='CDF', color='blue')
plt.scatter(x, cdf_values, color='red', zorder=5)

plt.title("CDF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.xticks(x)
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#7. Generate a continuous uniform distribution using NumPy and visualize it
import numpy as np
import matplotlib.pyplot as plt

def visualize_continuous_uniform(a=0, b=1, size=1000, bins=20):
    """
    Generate samples from a continuous uniform distribution and visualize them.

    Parameters:
    a (float): Lower bound of the distribution
    b (float): Upper bound of the distribution
    size (int): Number of random samples to generate
    bins (int): Number of histogram bins
    """
    # Generate random samples
    samples = np.random.uniform(a, b, size)

    # Compute theoretical PDF values
    x = np.linspace(a, b, 200)
    pdf = np.full_like(x, 1/(b - a))

    # Plot histogram of sampled data
    plt.hist(samples, bins=bins, density=True, alpha=0.6, edgecolor='black', label='Sample Histogram')

    # Overlay theoretical PDF
    plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

    # Beautify plot
    plt.title(f'Continuous Uniform Distribution (a={a}, b={b})')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Example usage:
    visualize_continuous_uniform(a=0, b=1, size=1000, bins=20)
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from the dataset and plot them.

    Parameters:
    data (list or np.array): Input numerical data

    Returns:
    np.array: Array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot 1: Line plot of Z-scores
    plt.figure()
    plt.plot(z_scores, marker='o', linestyle='-')
    plt.axhline(y=0, linestyle='--')
    plt.title('Z-Scores Line Plot')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.grid(True)
    plt.show()

    # Plot 2: Histogram of Z-scores
    plt.figure()
    plt.hist(z_scores, bins=10, edgecolor='black', density=True)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
if __name__ == "__main__":
    sample_data = [45, 50, 55, 60, 65, 70, 75, 80, 85]
    calculate_and_plot_z_scores(sample_data)
```

```python
#10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def demonstrate_clt(population_dist, sample_size=30, num_samples=1000):
    """
    Demonstrate the Central Limit Theorem for a given nonâ€‘normal population distribution.

    Parameters:
    -----------
    population_dist : callable
        Function that generates population data when called with `size` argument.
    sample_size : int
        Number of observations per sample.
    num_samples : int
        Number of samples to draw.
    """
    # 1) Generate a large â€œpopulationâ€ from the given distribution
    population = population_dist(size=100000)

    # 2) Draw many samples and record their means
    sample_means = [
        np.mean(np.random.choice(population, sample_size, replace=True))
        for _ in range(num_samples)
    ]

    # 3) Compute theoretical sampling distribution parameters
    pop_mean = np.mean(population)
    pop_std  = np.std(population)
    sampling_std = pop_std / np.sqrt(sample_size)

    # 4) Plot histogram of sample means
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, edgecolor='black', label='Sample means')

    # 5) Overlay corresponding normal PDF
    x = np.linspace(min(sample_means), max(sample_means), 200)
    plt.plot(x, norm.pdf(x, loc=pop_mean, scale=sampling_std), 'r-', lw=2, label='Normal PDF')

    # 6) Annotate
    plt.title(f'CLT Demonstration\nPopulation: Exponential, n={sample_size}, samples={num_samples}')
    plt.xlabel('Sample mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Example: use an exponential distribution with scale=2.0
    demonstrate_clt(
        population_dist=lambda size: np.random.exponential(scale=2.0, size=size),
        sample_size=30,
        num_samples=1000
    )
```

```python
#11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def verify_clt_with_normal_distribution(mu=50, sigma=10, sample_size=30, num_samples=1000):
    """
    Simulate samples from a normal distribution and verify the Central Limit Theorem (CLT).

    Parameters:
    mu (float): Mean of the normal distribution
    sigma (float): Standard deviation
    sample_size (int): Size of each sample
    num_samples (int): Number of samples to draw
    """
    # Generate sample means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
        sample_means.append(np.mean(sample))

    # Calculate theoretical mean and standard deviation of sample means
    sampling_mean = mu
    sampling_std = sigma / np.sqrt(sample_size)

    # Plot histogram of sample means
    plt.figure(figsize=(8, 5))
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, edgecolor='black', label='Sample Means')

    # Overlay the normal distribution curve
    x = np.linspace(min(sample_means), max(sample_means), 200)
    plt.plot(x, norm.pdf(x, loc=sampling_mean, scale=sampling_std), 'r-', lw=2, label='Theoretical Normal PDF')

    # Plot settings
    plt.title(f'CLT Verification with Normal Distribution\nÎ¼={mu}, Ïƒ={sigma}, n={sample_size}, samples={num_samples}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

# Run the program
if __name__ == "__main__":
    verify_clt_with_normal_distribution(mu=50, sigma=10, sample_size=30, num_samples=1000)
```

```python
#13.Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_binomial_distribution(n=10, p=0.5, size=1000):
    """
    Generate random variables and calculate their PMF using the Binomial distribution.

    Parameters:
    n (int): Number of trials
    p (float): Probability of success
    size (int): Number of random variables to generate
    """
    # Step 1: Generate random binomial samples
    samples = np.random.binomial(n=n, p=p, size=size)

    # Step 2: Get all possible outcomes (0 to n)
    x = np.arange(0, n+1)

    # Step 3: Calculate theoretical probabilities (PMF)
    pmf = binom.pmf(x, n, p)

    # Step 4: Plot PMF and Sample Histogram
    plt.figure(figsize=(8, 5))

    # Histogram of sampled data (normalized)
    plt.hist(samples, bins=np.arange(-0.5, n+1.5), density=True, alpha=0.6, edgecolor='black', label='Sampled Data')

    # Plot theoretical PMF
    plt.plot(x, pmf, 'ro-', label='Binomial PMF (theoretical)', linewidth=2)

    # Labels and legend
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(x)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Print probabilities
    print("Binomial PMF values:")
    for xi, prob in zip(x, pmf):
        print(f"P(X = {xi}) = {prob:.4f}")

# Example usage
simulate_binomial_distribution(n=10, p=0.5, size=1000)
```

```python
#14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    x (float): The data point
    mean (float): The population/sample mean
    std_dev (float): The population/sample standard deviation

    Returns:
    float: Z-score
    """
    return (x - mean) / std_dev

def plot_z_score(z):
    """
    Plot the standard normal distribution and shade the area up to the Z-score.

    Parameters:
    z (float): Z-score to mark on the standard normal curve
    """
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, loc=0, scale=1)

    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')

    # Shade area up to Z
    x_shade = np.linspace(-4, z, 500)
    y_shade = norm.pdf(x_shade)
    plt.fill_between(x_shade, y_shade, alpha=0.4, color='red', label=f'Z = {z:.2f}')

    plt.axvline(z, color='red', linestyle='--', linewidth=2)
    plt.title('Standard Normal Distribution with Z-Score')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    data_point = 72
    mean = 65
    std_dev = 5

    z = calculate_z_score(data_point, mean, std_dev)
    print(f"Z-score for data point {data_point} is: {z:.2f}")

    plot_z_score(z)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a Z-test for the population mean with known standard deviation.

    Parameters:
        sample_data (list or array): The sample dataset
        population_mean (float): Hypothesized population mean (H0)
        population_std (float): Known population standard deviation
        alpha (float): Significance level (default is 0.05)
        alternative (str): 'two-sided', 'greater', or 'less'

    Returns:
        dict: Test statistic, p-value, and conclusion
    """
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate p-value and decision rule
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
        reject = p_value < alpha
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
        reject = p_value < alpha
    elif alternative == 'less':
        p_value = norm.cdf(z_score)
        reject = p_value < alpha
    else:
        raise ValueError("Alternative must be 'two-sided', 'greater', or 'less'")

    return {
        'z_score': round(z_score, 3),
        'p_value': round(p_value, 5),
        'reject_null': reject,
        'decision': 'Reject Hâ‚€' if reject else 'Fail to Reject Hâ‚€'
    }

# Example usage
if __name__ == "__main__":
    sample = [52, 55, 50, 53, 54, 51, 56, 52, 54, 55]  # Sample dataset
    population_mean = 50       # Hâ‚€: Î¼ = 50
    population_std = 3         # Known standard deviation
    alpha = 0.05
    alternative = 'greater'    # Hâ‚: Î¼ > 50

    result = z_test(sample, population_mean, population_std, alpha, alternative)

    print("Z-Test Result:")
    print(f"Z-score      : {result['z_score']}")
    print(f"P-value      : {result['p_value']}")
    print(f"Decision     : {result['decision']}")
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

def compute_confidence_interval(data, confidence=0.95, population_std=None):
    """
    Compute the confidence interval for the mean of a dataset using Z-distribution.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (e.g., 0.95 for 95%)
        population_std (float): Population standard deviation (optional).
                                If None, use sample std (T-distribution recommended instead).

    Returns:
        (float, float): Lower and upper bounds of the confidence interval
    """
    data = np.array(data)
    n = len(data)
    sample_mean = np.mean(data)

    if population_std is None:
        raise ValueError("This function assumes population standard deviation is known.")

    z_critical = norm.ppf(1 - (1 - confidence) / 2)
    margin_of_error = z_critical * (population_std / np.sqrt(n))

    lower = sample_mean - margin_of_error
    upper = sample_mean + margin_of_error

    return round(lower, 2), round(upper, 2)

# Example usage
if __name__ == "__main__":
    sample_data = [48, 51, 50, 49, 52, 53, 50, 48, 49, 51]
    population_std = 2  # Known population standard deviation
    confidence_level = 0.95

    lower_bound, upper_bound = compute_confidence_interval(sample_data, confidence_level, population_std)

    print(f"95% Confidence Interval for the Mean: ({lower_bound}, {upper_bound})")
    print(f"Interpretation: We are 95% confident that the true population mean lies between {lower_bound} and {upper_bound}.")
```

```python
#18.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std_dev=1):
    """
    Calculate and plot the PDF of a normal distribution.

    Parameters:
        mean (float): Mean of the normal distribution
        std_dev (float): Standard deviation of the normal distribution
    """
    # Generate x values around the mean
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

    # Calculate PDF values
    pdf = norm.pdf(x, loc=mean, scale=std_dev)

    # Plot the PDF
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label=f'N({mean}, {std_dev}Â²)', color='blue')
    plt.title(f'Normal Distribution PDF\nMean = {mean}, Std Dev = {std_dev}')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axvline(mean, color='red', linestyle='--', label='Mean')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
plot_normal_pdf(mean=100, std_dev=15)
```

```python
#19.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def poisson_cdf_analysis(lam, k_max):
    """
    Calculate and interpret the CDF of a Poisson distribution up to k_max.

    Parameters:
        lam (float): Average rate (Î») of events
        k_max (int): Maximum number of events to calculate CDF up to
    """
    # Generate values from 0 to k_max
    k_values = np.arange(0, k_max + 1)

    # Calculate CDF values
    cdf_values = poisson.cdf(k_values, mu=lam)

    # Print CDF interpretation
    for k, cdf in zip(k_values, cdf_values):
        print(f"P(X â‰¤ {k}) = {cdf:.4f} â†’ Probability of â‰¤ {k} events occurring")

    # Plot the CDF
    plt.figure(figsize=(8, 5))
    plt.step(k_values, cdf_values, where='post', color='blue', label='Poisson CDF')
    plt.scatter(k_values, cdf_values, color='red')
    plt.title(f'Poisson CDF (Î» = {lam})')
    plt.xlabel('k (Number of Events)')
    plt.ylabel('Cumulative Probability P(X â‰¤ k)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
poisson_cdf_analysis(lam=3, k_max=10)
```

```python
#20. Simulate a random variable using a continuous uniform distribution and calculate its expected value
import numpy as np

def simulate_uniform_random_variable(a=10, b=20, size=10000):
    """
    Simulate a random variable from a continuous uniform distribution
    and calculate the expected value.

    Parameters:
        a (float): Lower bound of the distribution
        b (float): Upper bound of the distribution
        size (int): Number of random samples to generate

    Returns:
        dict: Simulated expected value and theoretical expected value
    """
    # Generate random samples from U(a, b)
    samples = np.random.uniform(low=a, high=b, size=size)

    # Calculate simulated expected value (mean of the sample)
    simulated_mean = np.mean(samples)

    # Theoretical expected value of U(a, b)
    theoretical_mean = (a + b) / 2

    return {
        'simulated_expected_value': round(simulated_mean, 4),
        'theoretical_expected_value': round(theoretical_mean, 4)
    }

# Example usage
result = simulate_uniform_random_variable(a=10, b=20, size=10000)

print("Simulated Expected Value:", result['simulated_expected_value'])
print("Theoretical Expected Value:", result['theoretical_expected_value'])
```

```python
#21.Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
        data1 (list or np.array): First dataset
        data2 (list or np.array): Second dataset
    """
    # Convert to numpy arrays
    data1 = np.array(data1)
    data2 = np.array(data2)

    # Calculate standard deviations
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    # Print results
    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")
    print("Interpretation: " + (
        "Dataset 1 is more spread out." if std1 > std2 else
        "Dataset 2 is more spread out." if std2 > std1 else
        "Both datasets have the same spread."
    ))

    # Plot boxplots for comparison
    plt.figure(figsize=(8, 5))
    plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'])
    plt.title('Comparison of Standard Deviations')
    plt.ylabel('Values')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Example usage
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=5, size=100)

compare_standard_deviations(data1, data2)
```

```python
#22.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np

def calculate_range_and_iqr(mu=100, sigma=15, size=1000):
    """
    Generate normal data and compute range and interquartile range (IQR).

    Parameters:
        mu (float): Mean of the normal distribution
        sigma (float): Standard deviation
        size (int): Number of data points

    Returns:
        dict: Range and IQR values
    """
    # Step 1: Generate the data
    data = np.random.normal(loc=mu, scale=sigma, size=size)

    # Step 2: Calculate range
    data_range = np.max(data) - np.min(data)

    # Step 3: Calculate IQR
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    return {
        'range': round(data_range, 2),
        'IQR': round(iqr, 2),
        'Q1': round(q1, 2),
        'Q3': round(q3, 2)
    }

# Example usage
result = calculate_range_and_iqr(mu=100, sigma=15, size=1000)

print(f"Range of Data      : {result['range']}")
print(f"Interquartile Range: {result['IQR']}")
print(f"Q1 (25th percentile): {result['Q1']}")
print(f"Q3 (75th percentile): {result['Q3']}")
```

```python
#23.Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def z_score_normalization_visual(data):
    """
    Apply Z-score normalization and visualize original vs normalized data.

    Parameters:
        data (list or array): Original dataset
    """
    data = np.array(data)

    # Step 1: Apply Z-score normalization
    normalized_data = zscore(data)

    # Step 2: Print original and normalized stats
    print("Original Data:")
    print(f"Mean: {np.mean(data):.2f}, Std Dev: {np.std(data, ddof=1):.2f}")
    print("\nAfter Z-Score Normalization:")
    print(f"Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data, ddof=1):.2f}")

    # Step 3: Plot comparison
    plt.figure(figsize=(10, 5))

    # Original data histogram
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=20, color='skyblue', edgecolor='black')
    plt.title("Original Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

    # Normalized data histogram
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=20, color='salmon', edgecolor='black')
    plt.title("Z-score Normalized Data")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# Example usage: Generate normally distributed data
np.random.seed(1)
data = np.random.normal(loc=50, scale=10, size=100)

z_score_normalization_visual(data)
```

```python
#24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mean=0, std_dev=1, size=1000):
    """
    Generate normal data and compute its skewness and kurtosis.

    Parameters:
        mean (float): Mean of the normal distribution
        std_dev (float): Standard deviation of the distribution
        size (int): Number of data points to generate

    Returns:
        dict: Skewness and kurtosis values
    """
    # Step 1: Generate normal data
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Step 2: Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data)  # Excess kurtosis (normal = 0)

    # Print interpretations
    print(f"Skewness: {skew_val:.4f} â†’ {'Symmetric' if abs(skew_val) < 0.1 else 'Skewed'}")
    print(f"Kurtosis: {kurt_val:.4f} â†’ {'Mesokurtic (normal)' if abs(kurt_val) < 0.1 else 'Leptokurtic' if kurt_val > 0 else 'Platykurtic'}")

    return {
        'skewness': round(skew_val, 4),
        'kurtosis': round(kurt_val, 4)
    }

# Example usage
result = calculate_skewness_kurtosis(mean=0, std_dev=1, size=1000)
```
